# Reading Carboxylate molecules from local CHEMBL database

```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import DataStructs
from rdkit.Chem import SaltRemover
from rdkit.Chem import PandasTools
from rdkit.Chem.Draw import IPythonConsole
import psycopg2
import pandas
import pandas.io.sql as psql
```

From Daylight examples, a SMARTS query for carboxylic acid is `[CX3](=O)[OX2H1]`

For the carboxylate form `[CX3](=O)[O-]`

For either: `[CX3](=O)[OX1H0-,OX2H1]`

```python
conn = psycopg2.connect(database='chembl_19',user='hjkulik',password='')
```

## Query for carboxylates into a pandas dataframe

Build a query for carboxylate containing molecules in CHEMBL with fewer than 20 atoms

```python
conn.rollback()
df = psql.read_sql_query("""SELECT molregno,m smiles,mol_numatoms(m) FROM rdk.mols WHERE m @> '[CX3](=O)[O-]'::qmol AND 
                           mol_numatoms(m) < 25 LIMIT 10000""", con=conn)
df.describe()
```

... and carboxylic acids

```python
conn.rollback()
df = psql.read_sql_query("""SELECT molregno,m smiles,mol_numatoms(m) FROM rdk.mols WHERE m @> '[CX3](=O)[OX2H1]'::qmol AND 
                           mol_numatoms(m) < 25 LIMIT 10000""", con=conn)
df.describe()
```

... and to make sure they both sum to the total... (actually, don't expect them to... could have both in same molecule)

```python
conn.rollback()
df = psql.read_sql("""SELECT molregno,m smiles,mol_numatoms(m),mol_send(m),bfp_to_binary_text(morganbv_fp(m)) 
                            FROM rdk.mols 
                            WHERE m @> '[CX3](=O)[OX1H0-,OX2H1]'::qmol 
                            AND mol_numatoms(m) < 25 LIMIT 10000""", con=conn)
df.describe()
```

```python
df.info()
```

Massage some of the data into usable rdkit format.

```python
df['mol_send']=df['mol_send'].map(lambda x: Chem.Mol(str(x)))
df['mbv_fp']=df['bfp_to_binary_text'].map(lambda x: DataStructs.CreateFromBinaryText(str(x)))
#df.to_pickle('CHEMBL_carboxyl.pkl')
```

```python
df.head()
```

## Strip some common ions out of the molecule objects

```python
remover = SaltRemover.SaltRemover(defnData="[Na,K,Li,Mg,Cl,Br,F]")
```

```python
df['mol_strip']=df['mol_send'].map(remover.StripMol)
```

```python
df.to_pickle('CHEMBL_carboxyl_strip.pkl')
```

## Filter out any molecules containing non-[H,N,C,O] elements

## Extracting fingerprints for building similarity matrix

Following the example here: http://nbviewer.ipython.org/gist/greglandrum/4316434
* First need to dump fingerprints (which kind?) from dataframe (values?) 
 * Lots of available fingerprints in the cartridge. Example uses Morgan Bit Vect: `morganbv_fp(mol,int default 2)`
* Calculate distance matrix
 * example uses "BulkTanimotoSimilarity"
 * also could use Metric Matrix Module http://www.rdkit.org/Python_Docs/rdkit.DataManip.Metric.rdMetricMatrixCalc-module.html
* Send distance matrix to diversity picker (rdSimDivPickers)

```python
from rdkit.DataManip.Metric.rdMetricMatrixCalc import GetTanimotoDistMat
dm = GetTanimotoDistMat(dffr.mbv_fp.tolist())
```

```python
from rdkit.SimDivFilters import rdSimDivPickers
```

```python
picker = rdSimDivPickers.MaxMinPicker()
```

```python
ids = picker.Pick(dm,dffr.shape[0],20)
```
